Univariate analysis (analisis univariat) adalah tahap analisis data yang fokus pada satu variabel (kolom) dalam satu waktu. Kita bisa menggunakan beberapa opsi untuk mendeskripsikan data univariat, antara lain:

```python
#explorasi Kolom Rating text
df["Rating text"].unique()
```

```python
#. Visualizing Categorical Data:
# Create a bar plot for a categorical variable
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='Rating text', data=df)
plt.title('Distribution of Rating Text Column')
plt.show()
```

### **4. BIVARIATE ANALYSIS**

 Analisis bivariat melibatkan analisis dua variabel secara bersamaan untuk menentukan apakah terdapat hubungan atau asosiasi di antara keduanya.

### **a. Numeric vs Numeric**

```python
# Pastikan tipe data numerik benar
df['Price range'] = pd.to_numeric(df['Price range'], errors='coerce')
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Price range', y='Aggregate rating', data=df, alpha=0.6, color='teal')
plt.title('Hubungan antara Price Range dan Aggregate rating')
plt.xlabel('Price Range')
plt.ylabel('Aggregate rating')
plt.show()
```

### **b. Kategorikal vs Numeric**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Has Online delivery', y='Aggregate rating', data=df, palette='Set2', hue='Has Online delivery', )
plt.title('Perbandingan Rating Berdasarkan Online Delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Aggregate Rating')
plt.show()
```

### **c. Kategorical vs Categorical**

```python
cross_tab = pd.crosstab(df['Has Table booking'], df['Has Online delivery'])
print("\n=== Cross Tab: Table Booking vs Online Delivery ===")
print(cross_tab)

sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Hubungan antara Fasilitas Table Booking dan Online Delivery')
plt.show()
```

```python
df = df[df['Has Table booking'].isin(['Yes', 'No'])]
```

```python
cross_tab = pd.crosstab(df['Has Table booking'], df['Has Online delivery'])
print("\n=== Cross Tab: Table Booking vs Online Delivery ===")
print(cross_tab)

sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Hubungan antara Fasilitas Table Booking dan Online Delivery')
plt.show()
```

### **5. Analisis Multivariate**

analisis multivariat, merupakan analisis yang melibatkan lebih dari 2 variabel

```python
#BOX PLOT 3 VARIABEL (Kota, Online Delivery, Rating)
# ===============================
#cut data 5 kota teratas
top_cities = df['City'].value_counts().nlargest(5).index
subset = df[df['City'].isin(top_cities)]
#==============================
plt.figure(figsize=(10,6))
sns.boxplot(
    data=subset,
    x='City',
    y='Aggregate rating',
    hue='Has Online delivery',
    palette='Set2'
)
plt.title('Perbandingan Rating Berdasarkan Online Delivery di 5 Kota Teratas')
plt.xlabel('City')
plt.ylabel('Aggregate Rating')
plt.legend(title='Online Delivery')
plt.show()
```

```python
plt.figure(figsize=(6,4))
numeric_cols = ['Average Cost for two','Votes','Aggregate rating','Price range']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='YlOrBr')
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```
