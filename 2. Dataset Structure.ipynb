# **Dataset Structure**
<p align="justify"> Dataset structure merujuk pada bagaimana data <b>disusun</b>, <b>diorganisasikan</b>, dan <b>direpresentasikan</b> agar dapat dipahami serta diolah oleh manusia maupun mesin. Struktur ini menjadi fondasi penting karena menentukan kemudahan analisis, konsistensi, serta efisiensi dalam proses <i>machine learning</i> maupun penelitian.</p>

<p align="justify"> Pada dataset tabular, struktur biasanya mengikuti format tabel dengan baris yang mewakili satu entitas atau sampel, sedangkan kolom menggambarkan karakteristik atau <i>variabe</i></p>

<p align="justify"> Untuk dataset berbasis gambar, audio, atau teks, struktur biasanya tidak lagi berbentuk tabel sederhana. Dataset gambar, misalnya, umumnya disimpan dalam folder yang sudah terorganisasi berdasarkan kelas atau kategori. Setiap folder berisi kumpulan file citra yang merepresentasikan label tertentu. Dataset teks dapat berbentuk dokumen dengan anotasi kategori, sedangkan dataset audio biasanya dilengkapi dengan metadata tambahan seperti durasi, format, dan transkrip. Selain itu, struktur dataset juga sering dilengkapi dengan metadata yang mendeskripsikan konteks data, sumber, dan skema anotasi. Metadata membantu peneliti memahami cara interpretasi data, misalnya informasi tentang format file, cara pengumpulan data, atau distribusi kelas.</p>

# **2. Tuple**

<p align="justify">
<b>Tuple</b>  adalah urutan objek Python yang <b>tidak berubah (immutable)</b>. Tuple adalah urutan, seperti daftar. Tuple juga memiliki index. Perbedaan utama antara tuple dan daftarnya adalah bahwa tuple tidak dapat diubah tidak seperti List Python. Tuple menggunakan <b>tanda kurung</b>, sedangkan List Python menggunakan tanda kurung siku.</p>

<p align="justify">Membuat tuple semudah memasukkan nilai-nilai yang dipisahkan koma. Secara opsional, Anda dapat memasukkan nilai-nilai yang dipisahkan koma ini di antara tanda kurung juga. Sebagai contoh :</p>

```python
#Contoh sederhana pembuatan tuple pada bahasa pemrograman python
tup1 = ('BigData & Data Mining', 'BigData & Predictive Analytic', 2022, 2022)
tup2 = (1, 2, 3, 4, 5 )
tup3 = ("a", "b", "c", "d")
```

```python
tup1
```

```python
tup1[1]
```

```python
tup2[2:6]
```

```python
#Cara mengakses nilai tuple
print ("tup1[0]: ", tup1[0])
print ("tup2[1:5]: ", tup2[1:5])
```

```python
tup2[1:5]
```

```python
#Tampilkan tup2 dan tup3 dalam satu baris
print (tup2,' & ',tup3)
print (tup2+tup3)
```

# **3. Dictionary**
Dictionary Python berbeda dengan List ataupun Tuple. Karena setiap urutanya berisi **key dan value**. Setiap key dipisahkan dari value-nya oleh **titik dua** (**:**), item dipisahkan oleh koma, dan semuanya tertutup dalam kurung kurawal **{}**. Dictionary kosong ditulis hanya dengan dua kurung kurawal, seperti ini: {}.

```python
dict = {'Name': 'Zara', 'Age': 20, 'Class': 'BigData1'}
```

```python
dict
```

```python
dict['Name']
```

```python
dict['Age']=8
```

```python
dict
```

```python
dict['University']="Amikom"
```

```python
dict
```

```python
#Update dictionary python
dict['Age'] = 8; # Mengubah entri yang sudah ada
dict['University'] = "Amikom" # Menambah entri baru

print ("dict['Age']: ", dict['Age'])
print ("dict['University']: ", dict['University'])
```

```python
#Tambahkan satu dictionary baru dengan Identitas nama, umur, dan kelas masing-masing
#pada dictionary dict

#Dictionary dict nantinya akan menjadi bentuk List
print('Cara 1 : ')
dict = [dict,{'Name': 'Tatang', 'Age': 11, 'Class': 'BigData0'}]
print(dict)

print('\nCara 2 : ')
dict = [{'Name': 'Zara', 'Age': 20, 'Class': 'BigData1'},
        {'Name': 'Tatang', 'Age': 11, 'Class': 'BigData0'}]
print(dict)
```

# **DataFrame dari List & Dictionary**

```python
#From dict of ndarrays / lists
import pandas as pd

data1 = {'buah': ['apel','pisang','jeruk','nanas','durian'],
        'sayur' : ['bayam','sawi','wortel','kentang','kacang'],
        'hewan': ['kucing','kambing','sapi','kelinci','buaya']}

df = pd.DataFrame(data1)

```

```python
df
```

```python
#Tampilkan Sapi
print(df['hewan'][2])
```

```python
#Tampilkan Kentang dan Buaya dalam satu baris
print(df['sayur'][3],' ',df['hewan'][4])
```

```python
#From a list of dicts
data2 = [{'Name': 'Zara', 'Age': 20, 'Class': 'BigData1'}, # index ke-0
        {'Name': 'lala', 'Age': 21, 'Class': 'BigData2'}, # index ke-1
        {'Name': 'lili', 'Age': 22, 'Class': 'BigData3'}] # index ke-3

df2 = pd.DataFrame(data2)

df2
```
